Here we start our data cleaning process where we clean our dataset which contain only one type of unique values 

```python
non_unique_list = []
for column in data.columns:
    print (column,data[column].nunique())
    if data[column].nunique() <2:
        non_unique_list.append(column)
```

Here we created a list which contains non unique values columns so we can directly drop the whole columns

```python
data.drop(columns= non_unique_list, inplace= True)
```

```python
columns = data.columns
id_vars = columns[:12]
value_vars = columns[12:]

print(id_vars)
print(value_vars)
```

```python
df_melt = pd.melt(data,
    id_vars=id_vars,
    value_vars= value_vars,
    var_name='Year',
    value_name='Value'
)
```

```python
df_melt['COMP_BREAKDOWN_1_LABEL'] = df_melt['COMP_BREAKDOWN_1_LABEL'].str.replace('Counterpart Area: ', '')
```

Counterpart Area is a common word which use in every records of columns COMP_BREAKDOWN_1_LABEL

```python
df_melt.drop(columns=['INDICATOR' , 'UNIT_MEASURE' , 'COMP_BREAKDOWN_1' , 'COMP_BREAKDOWN_2' , 'COMP_BREAKDOWN_3', 'REF_AREA'], errors='ignore', inplace = True)
```

Here we drop unwanted columns in our dataset

```python
df_melt.rename(columns ={"COMP_BREAKDOWN_1_LABEL" : "Counter_Part" , "COMP_BREAKDOWN_2_LABEL" : "Nature" , "COMP_BREAKDOWN_3_LABEL" : "Type"}, inplace =True)
```
